```python
import matplotlib.pyplot as plt

# Model names and their corresponding accuracies
models = ['LSTM', 'CNN', 'Bi-LSTM', 'Bi-LSTM(2-layer)', 'Hybrid-(CNN-LSTM)']
accuracies = [88.05, 89.02, 87.63, 88.52, 88.30]

# Create a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(models, accuracies, color='skyblue', edgecolor='black')

# Annotate each bar with its accuracy value
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height:.2f}%', ha='center', va='bottom', fontsize=10)

# Set chart title and labels
plt.title('Model Accuracy Comparison', fontsize=14)
plt.xlabel('Model', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.ylim(86.5, 90)  # Set y-axis limits for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
